```python
target_name = "FERNANDEZ"   # YOUR SURNAME (UPPERCASE)
SEED_NUM = "5"              # Last digit of student ID (string)

redacted_name = ""

for ch in target_name:
    if ch in "AEIOU":
        redacted_name += SEED_NUM
    else:
        redacted_name += ch

print("Original:", target_name)
print("Seed Num:", SEED_NUM)
print("Redacted:", redacted_name)
```

```python
first_letter = "F"          # First letter of your name
ASCII = ord(first_letter)

SEED_NUM = 5                # Integer now
loops = len("FERNANDEZ")    # Number of iterations

verification_hash = (ASCII * SEED_NUM) - loops

print("ASCII:", ASCII)
print("Loops:", loops)
print("Verification Hash:", verification_hash)
```

```python
correct_password = "python123"
attempts = 0

while True:
    user_input = input("Enter password: ")
    attempts += 1

    if user_input == correct_password:
        print("Access granted ✅")
        break
    else:
        print("Wrong password ❌")

print("Attempts before success:", attempts)
```
